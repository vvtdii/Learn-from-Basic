# Menunjukkan Perbandingan

## Bar Chart

Bar Chart merupakan grafik yang digambarkan dengan baratau batang vertikal.

Bar Chart menjelaskan jarak perdagangan dari suatu komoditas atau efek dalam periode tertentu.

## Line Chart

Line chart adalah grafik yang paling sederhana yang berbentuk garis.

Line chart  digunakan untuk melihat pola data time series, seperti trend dan musiman data. Menyajikan data yang berkesinambungan/kontinu, seperti suhu badan pasien rumah sakit, curah hujan per
bulan.

```python
data = {'year': [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010],
        'unemployment_rate 1': [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3],
        'unemployment_rate 2': [7.8, 6, 5.3, 7.1, 8.4, 7.3, 6, 6.8, 7.5, 8.2]
       }

df = pd.DataFrame(data)
```

```python
#Single line chart

import matplotlib.pyplot as plt

plt.plot(df['year'], df['unemployment_rate 1'], color='red', marker='o')
plt.title('unemployment rate vs year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('unemployment rate', fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Multi line chart

plt.plot(df['year'], df['unemployment_rate 1'], color='red', marker='o')
plt.plot(df['year'], df['unemployment_rate 2'], color='blue', marker='o')
plt.title('unemployment rate vs year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('unemployment rate', fontsize=14)
plt.grid(True)
plt.show()
```

# Menunjukkan Komposisi

## Pie Chart

Diagram yang berbentuk lingkaran.

Diagram ini digunakan untuk membandingan suatu bagian terhadap seluruh bagian. Pada pie chart satuan yang digunakan yaitu persentase.

```python
#Simple pie chart

plt.pie(train['Type'].value_counts(),autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Exploding pie chart

explode = (0.5,0)

plt.pie(train['Type'].value_counts(),explode=explode, radius=2, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Donut Pie Chart

plt.pie(train['Type'].value_counts(),autopct='%1.1f%%')

circle=plt.Circle((0,0), 0.4, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
```

```python
#Modified pie chart

import plotly.express as px

values=train['Type'].value_counts()
names=train['Type'].unique().tolist()

fig = px.pie(train, values=values, names=names, title='Type')

fig.show()
print(values)
```

## Waterfall Chart

Waterfall chart adalah salah satu grafik yang sering digunakan untuk menilai pencapaian target dari suatu variabel. Grafik ini menggambarkan bagaimana peningkatan atau penuruan nilai melalui serangkaian perubahan.

Sangat berguna dalam akuntansi dan analisis kualitatif.

Waterfall chart efektif digunakan untuk menampilkan hasil kumulatif dari data yang bernilai positif maupun negatif.

```python
import plotly.graph_objects as go

fig = go.Figure(go.Waterfall(
    x = [["2016", "2017", "2017", "2017", "2017", "2018", "2018", "2018", "2018"],
       ["initial", "q1", "q2", "q3", "total", "q1", "q2", "q3", "total"]],
    measure = ["absolute", "relative", "relative", "relative", "total", "relative", "relative", "relative", "total"],
    y = [10, 20, 30, -10, None, 10, 20, -40, None], base = 300,
    decreasing = {"marker":{"color":"Maroon", "line":{"color":"red", "width":2}}},
    increasing = {"marker":{"color":"Teal"}},
    totals = {"marker":{"color":"deep sky blue", "line":{"color":"blue", "width":3}}}
))

fig.update_layout(title = "Profit and loss statement", waterfallgap = 0.3)

fig.show()
```

# Menunjukkan Hubungan

## Scatter Plot

Scatter plot berbentuk titik-titik yang mewakili nilai untuk dua variabel numerik berbeda. Posisi setiap titik pada sumbu horizontal dan vertikal menunjukkan nilai untuk titik data individual.

Scatter plot berfungsi untuk mengamati hubungan antar variabel.

```python
plt.scatter(train['Reviews'],train['Rating']) #x=reviews, y=rating
plt.show()
```

```python
plt.scatter(train['Price'],train['Rating']) #x=price, y=rating
plt.show()
```

## Heatmap

Heatmap adalah visualisasi atau pemetaan dengan menampilkan data dengan representasi warna yang berbeda-beda. Biasanya, semakin tinggi angka suatu kelompok data, warnanya akan semakin gelap, umumnya disimbolkan dengan warna merah.

Heatmap adalah salah satu alat visualisasi terbaik untuk data poin yang padat. Heatmaps digunakan untuk memudahakan dalam pengidentifikasian cluster dimana ada konsentrasi tinggi suatu aktifitas. Heatmap juga berguna dalam cluster analysis atau hotspot analysis.

```python
%matplotlib inline
num = ["Rating","Reviews","Price"]
correlation = train[num].corr()
sns.heatmap(correlation,annot=True)
```
